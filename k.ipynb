```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
df=pd.read_csv("./amazon (1).csv",encoding="latin")
```

```python
sns.displot(x=df["number"],hue=df["month"])
```

```python
sns.stripplot(x=df["month"],hue=df["number"],y=df["state"])
```

```python
sns.relplot(y=df["number"],x=df["date"],kind="line")
plt.xticks(rotation=90)
```

```python
sns.swarmplot(x=df["state"],y=df["number"])
plt.xticks(rotation=90)
```

```python
sns.distplot(x=df["number"])
```

```python
sns.histplot(x=df["year"],y=df["number"])
```

```python
sns.barplot(x=df["state"],y=df["number"])
plt.xticks(rotation=90)
```

```python
sns.violinplot(x=df["month"],y=df["number"])
plt.xticks(rotation=90)
```

```python
a=df.groupby("year").sum().sort_values("number")

plt.pie(a['number'],labels=a.index,autopct="%0.2f",explode=[0.1]*20)
plt.pie([1],radius=0.5)
```

```python
sns.relplot(x=df["month"],y=df["number"],hue=df["state"],kind='line')
plt.xticks(rotation=90)
```

```python
sns.kdeplot(x=df["year"],y=df["number"],)
```

```python
sns.boxenplot(x=df["year"],y=df["number"],)
plt.xticks(rotation=90)
```

```python
sns.relplot(x=df["year"],y=df["number"],hue=df["month"],kind='line')
```

```python
df["state"].value_counts().plot(kind='pie')
plt.pie([1],colors="white",radius=0.5)
```

```python

sns.heatmap(df.pivot_table(index="state",columns="month",values="number",aggfunc='sum'))
```
