```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression()
x=df.iloc[:,0:11].values
y=df.iloc[:,12].values
from sklearn.preprocessing import MinMaxScaler
r=MinMaxScaler()
x=r.fit_transform(x)

from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(x,y,random_state=1)
mod.fit(xtr,ytr)
h=mod.predict(xts)
from sklearn.metrics import accuracy_score
accuracy_score(h,yts)


#This model can be used for predicting the color of the wine with 98% accuracy

```
