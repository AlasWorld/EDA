```python
d=[]
sub=["male","female"]
for i in df["gender"].unique():
  s=[]
  p=df[df["gender"]==i]
  s.append(p["math score"].mean())
  s.append(p["reading score"].mean())
  s.append(p["writing score"].mean())
  d.append(s)
plt.plot(sub,d)
# orange=reading score
# green =writing score
# blue= math score 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
prod=LogisticRegression()

from sklearn.metrics import accuracy_score

x=df.iloc[:,-3:]
for i in range (0,5):
  y=df.iloc[:,i]
  xtr,xts,ytr,yts = train_test_split(x,y, random_state=0)
  prod.fit(xtr,ytr)
  
  g=prod.predict(xts)
  print(accuracy_score(g,yts))

 


```

```python
y=df.iloc[:,0]
x=df.iloc[:,-3:]
xtr,xts,ytr,yts = train_test_split(x,y,random_state=0)
prod.fit(xtr,ytr)
  
g=prod.predict(xts)
print(accuracy_score(g,yts))
 
prod.coef_
# so the  maths marks are  correlated here with positive values, reading marks, writing marks  are the values
# which are negatively correlated.

```
