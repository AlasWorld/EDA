```python
from sklearn.datasets import load_iris
iris=load_iris()
from sklearn.ensemble import RandomForestClassifier
xtr,xts,ytr,yts=sklearn.model_selection.train_test_split(iris.data,iris.target)
clasff=RandomForestClassifier(n_estimators=2,min_samples_leaf=3,min_samples_split=2)
clasff.fit(xtr,ytr)

sklearn.metrics.accuracy_score(yts,clasff.predict(xts))


```

```python
from sklearn.model_selection import GridSearchCV


s=GridSearchCV(estimator=clasff,param_grid={"n_estimators":[2,3,4],"min_samples_split":[3,4],"min_samples_leaf":[3,5,6]})
s.fit(xtr,ytr)
s.predict(xts)
sklearn.metrics.accuracy_score(s.predict(xts),yts)
s.best_params_
```
